```python
plt.figure(1,figsize=(10, 8))
plt.subplots_adjust(hspace=0.4, wspace=0.7)

plt.subplot(2, 2, 1)
sns.boxplot(df['sepal_length'],color='r')
plt.title('sepal_length')

plt.subplot(2, 2, 2)
sns.boxplot(df['sepal_width'],color='g')
plt.title('sepal_width')

plt.subplot(2, 2, 3)
sns.boxplot(df['petal_length'],color='orange')
plt.title('petal_length')

plt.subplot(2, 2, 4)
sns.boxplot(df['petal_width'],color='grey')
plt.title('petal_width')
plt.show()
```
